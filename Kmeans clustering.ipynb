```r
Timeuse <- read.csv(
    file='atus_df_weekend_prop.csv',
    sep=',',
    dec='.',
    header=TRUE,
    row.names=1
)

head(Timeuse)
dim(Timeuse)
```

```r
time_lst <- c('sleeping', 'household', 'caringHHmem', 'caringNHHmem',
            'working', 'education', 'consumer', 'personalcare', 'householdservice', 
            'eating', 'socializing', 'sports', 'religious', 'volunteer')
```

```r
model <- kmeans(Timeuse[, time_lst] , algorithm="Lloyd", centers = 8, nstart=25)
```

```r
Timeuse_cluster <- data.frame(Timeuse,
  cluster = as.factor(model$cluster)
)

head(Timeuse_cluster)
```

```r
(BSS <- model$betweenss)
(TSS <- model$totss)
BSS / TSS * 100
```

```r
library(ggplot2)
#library(ggrepel)
library(factoextra)
fviz_cluster(model, data = Timeuse)
```

```r
library(parameters)
res_kmeans <- cluster_analysis(Timeuse[, time_lst], n=8, method="kmeans")
predict(res_kmeans)
```
